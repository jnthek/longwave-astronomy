```python
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
```

```python
data, samplerate = sf.read('test.wav') 
f,t,Zxx = signal.stft(data[:,0], fs=samplerate, nperseg=samplerate)
print (Zxx.shape)
print (samplerate)
```

```python
ind_low = int(0*len(f)/np.max(f))
ind_high = int(24000*len(f)/np.max(f))
```

```python
plt.figure(figsize=(16,9))
plt.title("White noise recording from blank FM channel (~97.4 MHz)")
plt.imshow(np.log10(np.abs(Zxx[ind_low:ind_high,:].T)),aspect='auto', cmap='jet')
plt.xlabel("Freq (kHz)")
plt.ylabel("Time index")
plt.savefig("FFT_waterfall.png", dpi=100, bbox_inches="tight", transparent=False)
```

```python
plt.figure(figsize=(16,9))
plt.title("White noise recording from blank FM channel (~97.4 MHz)")
plt.semilogy(f[ind_low:ind_high]/1e3, np.mean(np.abs(Zxx[ind_low:ind_high,:]),axis=1), color='red')
plt.xlim([0,24])
plt.grid()
plt.xlabel("Freq (kHz)")
plt.ylabel("Amplitude")
plt.savefig("FFT_power.png", dpi=100, bbox_inches="tight", transparent=False)
```
