```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Last one is the background c.f. https://solar-radio.gsfc.nasa.gov/wind/one_minute_doc.html
bg_spec_no_ant = np.loadtxt("19941111.R2")[:,1440] 
bg_spec_ant = np.loadtxt("19941117.R2")[:,1440] 

flow=1.075
fhigh=13.825
freq = np.linspace(flow, fhigh, 256)
```

```python
plt.figure(figsize=(16,9))
# plt.loglog(freq, bg_spec_no_ant)
plt.plot(freq, bg_spec_ant)

plt.figure(figsize=(16,9))
plt.semilogx(freq, bg_spec_ant-bg_spec_no_ant)
```
